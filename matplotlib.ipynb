```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]

gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot(years, gdp, color="green", marker='o', linestyle='solid')

plt.title('GPD Nominal')

plt.ylabel("Bilhoes de $")

plt.show()
```

```python
#Grafico em barras

movies = ["Annie Hall", "Ben-hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

#barras possuem o tamanho padrao de 0.8, entao adicionaremos 0.1 as coordenadas a esquerda para que cada barra seja
# centralizada

xs = [i + 0.1 for i, _ in enumerate(movies)] #Coordenadas a esquerda para que cada barra seja centralizada

plt.bar(xs, num_oscars)

plt.ylabel('# de premiacoes')
plt.title("Meus Filmes favoritos")

#Nomeia o eixo x com nomes de filmes na barra central

plt.xticks([i + 0.1 for i,  _ in enumerate(movies)], movies)

plt.show()
```

```python
#LIST COMPREHENSION
# Sintaxe: newLIst = [expression for item in interable if condition == true]

lista = ['1','14','33','55','12','556','3','261','21','212','421']
newlist = [x for x in lista if '1' in x]
print(newlist)

lista2 = [1,3,4,53,2,3,534,363, 21, 42, 214,1]
newlist2 = [y for y in lista2 if '1' in str(y)] # Se durante essa iteracao a expressao y for igual a 1, adicione na lista2

print(newlist2)
print(sum(newlist2))
```
